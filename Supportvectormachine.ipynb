```python
x=cancer.iloc[:,:2]
y  = cancer.iloc[:,-1]
```

```python
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
from sklearn.svm import SVC
```

```python
svcDiag=SVC(C=10,kernel="linear")
svcDiag.fit(x_train,y_train)
```

```python
pred = svcDiag.predict(x_test)
pred
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt

# Assuming svcDiag is your trained model and x_train, y_train are your features and labels
DecisionBoundaryDisplay.from_estimator(svcDiag, X=x_train, response_method="predict",
                                       cmap=plt.cm.Spectral, alpha=0.75)

# Scatter plot for the training points
plt.scatter(x_train.iloc[:, 0], x_train.iloc[:, 1], c=y_train, s=30, cmap=plt.cm.Spectral)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,pred)
acc
```

```python
svgDjagos = SVC(C=10,kernel="rbf",gamma=2)
svgDjagos.fit(x_train,y_train)
```

```python
svcpred = svgDjagos.predict(x_test)
svcpred
```

```python
ac = accuracy_score(svcpred,y_test)
ac
```

```python
svgDjagos.get_params()
```

```python
svcc = SVC(C=10,kernel="rbf",probability=True,gamma=2,random_state=5)
svcc.fit(x_train,y_train)
```

```python
svcprob = svcc.predict_proba(x_test)
svcprob
```

```python
pred1 = svcc.predict(x_test)
pred1
```

```python
score = accuracy_score(y_test,pred1)
score
```
